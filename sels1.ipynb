```python

import pandas as pd
df = pd.read_csv("F:/Jupyter_work/udemy_code/Sales Register 01.01.2016 - 05.08.2019.csv",encoding = 'unicode_escape') 

df['Date of Sale'] = pd.to_datetime(df['Date of Sale'],format='%d-%m-%y')
df['year']= df['Date of Sale'].dt.year
df['month']= df['Date of Sale'].dt.month

df["year"].apply(str)
df["year"]=df["year"].apply(str)
df["month"].apply(str)
df["month"]=df["month"].apply(str)

# df.columns
df1 = df[[ 'Origin', 'Date of Sale','Contract Qty', 'year', 'month']]
df1 = df1.drop(columns='Date of Sale')
df1
# df1.to_csv('df1.csv')
```

```python
df_grp = df1.groupby(['Origin','year','month']).sum()
df_grp.reset_index(level=0, inplace=True)
df_grp.reset_index(level=0, inplace=True)
df_grp.reset_index(level=0, inplace=True)
df_grp['Date'] = pd.to_datetime(df_grp.assign(Day=1).loc[:, ['year','month','Day']])
df_grp = df_grp.drop(columns=['month','year'])
df_grp
```

```python
df_grp.to_csv("Sorigin2.csv")
```
